```python
from typing_extensions import MutableMapping
#task3
import numpy as np
x = np.array([[1, 2],
              [3, 5]])
y = np.array([[5, 6],
              [7, 8]])
v = np.array([9, 10])
w = np.array([11, 12])
#1
add = x + y
print(add)
#2
sub= x - y
print(sub)
#3
multi = x * 7
print(multi)
#4
square = y ** 2
print(square)
#5
dot_1 = np.dot(v, w)
print(dot_1)
dot_2 = np.dot(x, v)
print(dot_2)
dot_3 = np.dot(x, y)
print(dot_3)
#6
concat_xy_rows = np.concatenate((x, y))
print(concat_xy_rows)
concat_vw_cols = np.concatenate((v.reshape(-1,1), w.reshape(-1,1)))
print(concat_vw_cols)
#7
# The original error was due to concatenating a 2D array `x` (shape (2,2)) with a 1D array `v` (shape (2,)).
# To fix this, `v` needs to be reshaped to a 2D array (e.g., (1,2)) to be compatible for concatenation along axis 0.
concat_xv=np.concatenate((x, v.reshape(1, -1)), axis=0)
print(concat_xv)
#NumPy cannot concatenate a 2D array with a 1D array unless dimensions match.
#To concatenate successfully, v must be reshaped.
```
